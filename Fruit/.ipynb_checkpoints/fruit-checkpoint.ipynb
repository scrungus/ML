```python
import pandas as pd
fruit = pd.read_table("fruit_data.txt")
fruit.head()
```

```python
y = fruit[["fruit_name"]]
X = fruit[["mass","width","height","color_score"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

sclr = StandardScaler()
X_train = sclr.fit_transform(X_train)
X_test = sclr.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import numpy as np

acc = []
k_val = []
con_m = []

for k in range(1,8):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, np.ravel(y_train))
    y_p = knn.predict(X_test)
    k_val.append(k)
    acc.append(metrics.accuracy_score(y_test,y_p)*100)
    con_m.append(metrics.confusion_matrix(y_test,y_p))
    print("k = ",k,":")
    print("Accuracy : {:.2f}".format(acc[k-1]),"%.")
    print("Confusion Matrix : ")
    for line in con_m[k-1]:
        print("   ","".join(str(line)))
    print()
    print("-----------------")
    print()
```

```python
from sklearn.naive_bayes import GaussianNB

fruit = pd.read_table("fruit_data.txt")
y = fruit[["fruit_name"]]
X = fruit[["mass","width","height","color_score"]]

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
sclr = StandardScaler()
X_train = sclr.fit_transform(X_train)
X_test = sclr.transform(X_test)

gb = GaussianNB()
gb.fit(X_train, np.ravel(y_train))
y_p = gb.predict(X_test)
print("Results : ")
print("Accuracy : {:.2f}".format(metrics.accuracy_score(y_test,y_p)*100),"%.")
print("Confusion Matrix : ")
for line in metrics.confusion_matrix(y_test,y_p):
    print("   ","".join(str(line)))
```

```python
fruit = pd.read_table("fruit_data.txt")
y = fruit[["fruit_name"]]
X = fruit[["mass","width","height","color_score"]]

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)

sclr = StandardScaler()
X_train = sclr.fit_transform(X_train)
X_test = sclr.transform(X_test)

prior = pd.value_counts(y_train['fruit_name']) / len(y_train)

means = []
means = np.array([X_train[np.where(y_train['fruit_name']==i)].mean(axis=0) for i in prior.index.values])
sds = np.array([X_train[np.where(y_train['fruit_name']==i)].std(axis=0) for i in prior.index.values])
```

```python
count = 0
y_p = []
for idd,i in enumerate(X_test):
    probs = []
    print("Item :",idd+1)
    
    for j in range(prior.index.size):
        result = ( 1 / np.sqrt( 2 * np.pi * sds[j] **2 )\
        * np.exp( -0.5 * (( i - means[j])\
        /sds[j]) **2 )).prod() * prior.values[j]
        probs.append(result)
        
    probs = probs / np.sum(probs)
    
    predicted_class = np.where(probs == np.amax(probs))
    
    y_p.append(prior.index.values[predicted_class[0].item(0)])
    
    print("Predicted : '"+prior.index.values[predicted_class[0].item(0)]+"'")
    print("Actual : '"+y_test.iloc[idd]['fruit_name']+"'")
    
    if prior.index.values[predicted_class[0].item(0)] == y_test.iloc[idd]['fruit_name']:
        count += 1
print()      
print("Accuracy : {:.2f}".format(metrics.accuracy_score(y_test,y_p)*100),"%.")
print("Confusion Matrix : ")
for line in metrics.confusion_matrix(y_test,y_p):
    print("   ","".join(str(line)))
```
